```python
def get_digit(n, d_index):
    '''d_index indexes the indices from 0 from the right-most digit'''
    return (n // 10**d_index) % 10
```

```python
input_nums = arr.array('i', [random.randint(1, 10000) for _ in range(20)])
input_nums
```

```python
def radix_sort(nums):
    print(nums)
    biggest = max(nums)
    num_digits = len(str(biggest))
    
    for dig_idx in range(num_digits):
        # calculate how many numbers belong to each bucket
        nums_per_digits = arr.array('i', [0 for _ in range(10)])
        for num in nums:
            nums_per_digits[get_digit(num, dig_idx)] += 1

        # generate prefix sum array
        prefix_sum = arr.array('i', [0 for _ in range(10)])
        for i in range(1, 10):
            prefix_sum[i] = prefix_sum[i-1] + nums_per_digits[i-1]

        # move numbers to bucket separated array
        bucketed_nums = arr.array('i', [random.randint(1, 10000) for _ in range(len(nums))])
        for num in nums:
            digit = get_digit(num, dig_idx)
            new_arr_idx = prefix_sum[digit]
            bucketed_nums[new_arr_idx] = num
            prefix_sum[digit] += 1

        nums = bucketed_nums
        
        print(nums)
```
